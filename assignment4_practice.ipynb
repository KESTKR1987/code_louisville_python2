```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import re
```

```python
nhl_df=pd.read_csv("nhl.csv")
cities=pd.read_html("wikipedia_data.html")[1]
cities=cities.iloc[:-1,[0,3,5,6,7,8]]
```

```python
nhl_df['team']=nhl_df['team'].str.replace('[\w.]*\ ', '')
nhl_df['team']=nhl_df['team'].str.replace('\*','')
cities['NHL']=cities['NHL'].str.replace('[\w.]*\ ', '')
cities['NHL']=cities['NHL'].str.replace('\*','')
nhlMask = nhl_df['team'] != 'Division'
nhl_df=nhl_df.where(nhlMask).dropna()
nhl_df['W']=pd.to_numeric(nhl_df['W'])
nhl_df['L']=pd.to_numeric(nhl_df['L'])
```

```python
is_2018 = nhl_df['year']==2018
is_2018.head()
```

```python
nhl_df_2018 = nhl_df[is_2018]
nhl_df_2018.shape
```

```python
nhl_df_2018['W/L Ratio'] = nhl_df_2018['W'] / nhl_df_2018['L']
nhl_df_2018.head()
```

```python
merged_nhl_df_2018 = pd.merge(nhl_df_2018, cities, how = 'inner', on = 'Metropolitan area')
merged2_nhl_df_2018 = merged_nhl_df_2018.set_index('Metropolitan area')

merged2_nhl_df_2018
```
