```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
#imports the discovery rate csv file
discovery_rate = pd.read_csv('Discovery rate.csv',
                            skiprows=1)
discovery_rate = discovery_rate.replace('...', np.NaN)
discovery_rate.columns = ['year', 'Cumulative Discovered Species', 'Discovered Species']
discovery_rate.head()
```

```python
discovery_rate_1964 = discovery_rate.query('year >= 1964')
discovery_rate_1964.head(57)
```

```python
plt.figure(figsize=(30,15))
plt.hist(discovery_rate_1964['year'], color='green', edgecolor='black', bins=int(57/1))

plt.title('Discovery Rate since 1964')
plt.xlabel('Year')
plt.ylabel('Number of Species')
```
